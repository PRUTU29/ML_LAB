```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split

# load the iris dataset
iris = datasets.load_iris()

df = pd.DataFrame(iris.data)
label = pd.DataFrame(iris.target)

# Let us split 70% for training & 30% for testing
x_train, x_test, y_train, y_test = train_test_split(
    df, label, test_size=0.3, shuffle=True
)

# x -> df, y -> label
x_train, x_val, y_train, y_val = train_test_split(
    x_train, y_train, test_size=0.3, shuffle=True
)

print("Total no. of samples in the dataset :", df.shape[0])
print("Total no. of labels in the dataset :", label.shape[0])

print("Total no. of samples in training set x_train :", x_train.shape[0])
print("Total no. of labels in training set y_train :", y_train.shape[0])

print("Total no. of samples in testing set x_test :", x_test.shape[0])
print("Total no. of labels in testing set y_test :", y_test.shape[0])

print("Total no. of samples in validation set x_val :", x_val.shape[0])
print("Total no. of labels in validation set y_val :", y_val.shape[0])

print("The 1st 3 samples of training set")
print(x_train.head(3))

print("The random 3 samples of training set")
print(x_train.sample(3))

print("The 1st 3 samples of testing set")
print(x_test.head(3))

print("The 3 samples of randomization")
print(x_test.sample(3))

print("The 1st 3 samples of validation set")
print(x_val.head(3))

print("The random 3 samples of validation")
print(x_val.sample(3))
```
